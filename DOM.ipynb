```python
from bs4 import BeautifulSoup 
import requests
from lxml import etree
## Para complementar: https://www.mclibre.org/consultar/xml/lecciones/xml-xpath.html 
```

```python
URL = "https://www.cnbc.com/world/?region=world"
page = requests.get(URL)
print(page)
```

```python
soup= BeautifulSoup(page.content)
print(soup)
```

```python
html= etree.HTML(str(soup))
html
```

```python
elemento2=html.xpath('//div[@id="HomePageInternational-riverPlus-11-0"]/div/div/div[@class="RiverHeadline-headline RiverHeadline-hasThumbnail"]/a') #//h2[@class='FeaturedCard-packagedCardTitle']/a
print(elemento2)
```
