```python
from bs4 import BeautifulSoup 
import requests
from lxml import etree
## Para complementar: https://www.mclibre.org/consultar/xml/lecciones/xml-xpath.html 
```

```python
URL = "https://www.reuters.com/"
page = requests.get(URL)
print(page)
```

```python
soup= BeautifulSoup(page.content)
print(soup)
```

```python
html= etree.HTML(str(soup))
html
```

```python
elemento=html.xpath("//div[@class='media-story-card__header__1NpsG']/a")
```

```python
elemento
```

```python
print(elemento[0].text)
```

```python
elemento2=html.xpath("//ul[@class='home-page-grid__home-hub__a11eg']//a[@data-testid='Heading']")
print(elemento2[0].text)
```

## TRABAJANDO CON Bsp

## Find
Procedemos a llevar el contenido de la pgina a formato Bsp4 y ubicar la etiqueta de inters. Para esto usaremos el mtodo Find de la Librera Bs4, este nos devolver el primer contenido que encuentre de la etiqueta (en este caso el primer "a").

```python
a=soup.find("a", attrs={"data-testid": "Heading"})
```

```python
a
```

```python
a.text
```

## Mtodo "find_all" 
Ahora que probamos que todo va bien con "find", buscaremos traer una lista de todos los a de la pgina. Para esto nos servir "find_all"

```python

a=soup.find_all("a", attrs={"data-testid": "Heading"})
len(a)
```

```python
todos_elementos_text=[item.text for item in a]
print(todos_elementos_text)
```

## Mtodo "find_parent" y "find_parents"

```python
text=soup.find(string='US lawmaker calls for ending Huawei, SMIC exports after chip breakthrough')
text.find_parent()
```

```python
text.find_parent('div')
```

```python
parents=text.find_parents("div", class_="text-story-card__basic__ITZwh")
print(parents)
```
