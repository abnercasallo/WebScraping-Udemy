# Función para scrapear la página

```python
def AmazonScrap():
    names=driver.find_elements(By.XPATH, "//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']")
    prices=driver.find_elements(By.XPATH, '//div[@class="a-section a-spacing-none a-spacing-top-small s-price-instructions-style"]')

    namesText=[]
    for i in range(0,len(names)):
        name=names[i].text
        namesText.append(name)
   
    pricesList=[]
    for i in range(0,len(prices)):
        price=prices[i].text
        pricesList.append(price)
    
    precios_extraidos = []

    # Itera sobre la lista y divide cada elemento en fragmentos basados en \\
    for elemento in pricesList:
        try:     
            fragmentos = elemento.split('\nUS')
            precios_extraidos.append(fragmentos[1])
        except:
            precios_extraidos.append("No tiene precio")
         
    precios_sin_pvpr = []

# Itera sobre la lista y quita "PVPR:" de cada elemento
    for elemento in precios_extraidos:
            precio_sin_pvpr=elemento.replace("\n", ".")
            precio_sin_pvpr = elemento.replace(' PVPR:', '')  # Reemplaza " PVPR:" con una cadena vacía
            precio_sin_pvpr = precio_sin_pvpr.replace(' Precio lista ed. digital:', '')
            precio_sin_pvpr = precio_sin_pvpr.replace(' Precio lista ed. impresa:', '')
            precio_sin_pvpr = precio_sin_pvpr.replace('Gratis con la membresía Kindle Unlimited Únete ahora', '')
            precio_sin_pvpr = precio_sin_pvpr.replace(' para alquilar', '')
            precio_sin_pvpr = precio_sin_pvpr.replace('Precio de preventa garantizado.', '')
            precio_sin_pvpr = precio_sin_pvpr.replace('\n', '.')
            precios_sin_pvpr.append(precio_sin_pvpr)

        
    df = pd.DataFrame({'Títulos': namesText, 'Precio':precios_sin_pvpr})
        
    return df
    
```
