## III. Seleccionar Menú Drop-Down y  Extracción de todas sus opciones: Método "find_element_by_name" y función"Select"

```python
#1.Función que extrae todas las opciones por el nombre del nodo ('distrito', 'provincia', 'departamentoAux', 'producto...)
#//*[@id="contact"]/div/div[1]/div/div[2]/select

from selenium.webdriver.support.select import Select

def elementos(x):
    lista=[]
    nodo=driver.find_element(By.NAME, x)     #Versión anterior: driver.find_element_by_name(x)
    lugares= Select(nodo) ##De la librería
    for lugar in lugares.options:
        lista.append(lugar.text)
    return lista
```

```python
#//select[@name="provincia"]
print(elementos('provincia'))
```

```python
##Previamente se debe seleccionar la provincia
print(elementos("distrito"))
```

```python
print(elementos("producto"))
```
