## Avanzamos las páginas usando Selenium
La estructura de las referencias es:
 1. "//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[2]/a"  :Página1
 2. "//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[3]/a"  :Página2
 3. "//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[4]/a"  :Página3
 ...
 
 Avancemos a la página2


```python
#'//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[3]/a'
buscar=driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[3]/a')
buscar.click()
```

```python
#//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[2]/a  :Página1
#//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[3]/a  :Página2:
# //*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[4]/a :Página3
#...
# //*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li[6]/a : página5


for i in range(3,6):
    buscar=driver.find_element(By.XPATH, '//*[@id="__next"]/div[2]/div[1]/div[2]/div/div[1]/div[6]/div[1]/div/ul/li['+str(i)+']/a')
    buscar.click()
    sleep(randint(3,5))
    for _ in range(8):  
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        sleep(randint(1,3))
    df2 = pd.read_html(driver.page_source, attrs = {'class': 'sc-b28ea1c6-3'})
    df2=df2[0]
    df1= pd.concat([df1, df2], ignore_index=True)

```
